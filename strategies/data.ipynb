```python
from openalgo import api

# Initialize the API client
client = api(api_key='your-openalgo-api-key', host='http://127.0.0.1:5000')

# Fetch historical data for SBIN
df = client.history(
    symbol="SBIN",
    exchange="NSE",
    interval="D",
    start_date="2010-01-01",
    end_date="2012-07-24"
)

df
```
